```python

from os import system
from matplotlib.pyplot import semilogy
from pylab import *
import spline
import numpy as np
from scipy import special
from scipy import interpolate
```

```python

def fn(w):
    return (w**(1+special.j0(w)))/(np.sqrt(1-w+(100*(w**2))-(100*(w**3))))

def fn_1(w):
    return (-1*(((w**special.j0(w))*((-100*(w**3))+(2*((100*(w**3))-(100*(w**2))+w-1)*special.j0(w))-((2*((100*(w**3))-(100*(w**2)) +w-1)*w*np.log(w)*special.j1(w))+w-2)))))/(2*(((-100*(w**3))+(100*(w**2)) -w+1)**1.5))

def deri1(x):
    y=0.0
    y=-(((x**special.j0(x))*(-100*x**3 + 2*(100*x**3 - 100*x**2 + x-1)*special.j0(x) - (2*(100*x**3 - 100*x**2 + x-1)*x*math.log10(x)*special.j1(x)) +x-2))/(2*(-100*x**3 + 100*x**2 -x +1)**1.5))
    return y
def splinenak(x,y,n):
    a=np.zeros(n)
    b=np.zeros(n)
    c=np.zeros(n)
    d=np.zeros(n)
    y2=np.zeros(n)


    b[0]=(x[1]-x[0])+(2*(x[2]-x[1]))
    c[0]=(x[2]-x[1])-(x[1]-x[0])
    p=((y[2]-y[1])/(x[2]-x[1]))-((y[1]-y[0])/(x[1]-x[0]))
    d[0]=(6*p*(x[2]-x[1]))/((x[2]-x[0]))


    a[n-1]=(x[n-2]-x[n-3])-(x[n-1]-x[n-2])
    b[n-1]=(2*(x[n-2]-x[n-3]))+(x[n-1]-x[n-2])
    p=((y[n-1]-y[n-2])/(x[n-1]-x[n-2]))-((y[n-2]-y[n-3])/(x[n-2]-x[n-3]))
    d[n-1]=(6*p*(x[n-2]-x[n-3]))/((x[n-1]-x[n-3]))


    for i in range(1,n-1):

        a[i]=(x[i]-x[i-1])
        c[i]=(x[i+1]-x[i])
        b[i]=2*(a[i]+c[i])
        d[i]=6*(((y[i+1]-y[i])/(x[i+1]-x[i]))-((y[i]-y[i-1])/(x[i]-x[i-1])))
    for i in range(1,n-1):
        p=a[i+1]*c[i]
        b[i+1]=b[i+1]-(p/(b[i]))
        p=a[i+1]*d[i]
        d[i+1]=d[i+1]-(p/(b[i]))

    y2[n-1]=d[n-1]/b[n-1]


    for k in range(n-2,-1,-1):
        y2[k]=(d[k]-(c[k]*y2[k+1]))/b[k]
    return y2
```

```python
xa=np.arange(0.1,0.9004,0.004)
xx=np.arange(0.1,0.9001,0.001)
fx=[fn(w) for w in xa]
fx_t=[fn(w) for w in xx]
d=[0.1,0.9]
fx_1=[deri1(w) for w in d]
y2a=spline.spline(xa,fx,fx_1[0],fx_1[1])
y2b=spline.splinenak(xa,fx)
y2c=spline.spline(xa,fx,2*10**30,2*10**30)
```
