```python
from pylab import *
from os import system
import spline
```

```python
xa=linspace(0,10,11)
ya=sin(xa)
y2a=zeros(ya.shape)
xx=linspace(-5,15,101)
yt=sin(xx)
yy=zeros(xx.shape)
u=zeros(ya.shape)
```

```python
y2a=spline.spline(xa,ya,1.0,cos(10.0))
figure(1)
plot(xa,ya,'r',xa,y2a,'g')
legend(["ya","y2a"])
grid(True)
show()
```

```python
y2a[0]=y2a[-1]=0
```

```python

yyb=spline.splintn(xa,ya,y2a,xx)
figure(2)
plot(xa,ya,'ro',xx,yt,'b+',xx,yyb,'k')
xlim([0,10])
ylim([-1,1])
grid(True)
show()
```

```python

figure(3)
semilogy(xx,abs(yt-yyb))
xlim([0,10])
grid(True)
show()
```

```python
yyb=spline.splintn(xa,ya,y2a,xx)
yyb[0]=yyb[10]=0
figure(2)
plot(xa,ya,'ro',xx,yt,'b+',xx,yyb,'k')
xlim([0,10])
ylim([-1,1])
grid(True)
show()
```

```python

figure(3)
semilogy(xx,abs(yt-yyb))
grid(True)
show()
```

```python
from pylab import *
from os import system
import spline
import numpy as np
from scipy import special
xa=np.arange(0.1,0.9,0.005)

fx=[(w**(1+special.j0(w)))/(np.sqrt(1-w+(100*(w**2))-(100*(w**3)))) for w in xa]
```

```python
figure(1)
plot(xa,fx,'r')
legend(["fx"])
grid(True)
show()
```
